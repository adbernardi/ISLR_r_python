```python
# Reading in the College data set 
# parts a and b

college = pd.read_csv('/Users/anthonybernardi/Desktop/Werk/ISLR/Data_Sets/College.csv', header=0, index_col=0)

college.head()
```

```python
# Fixing the first column and making it more legible
# Think we can actually do this with the index 
# done
# Getting a summary 
# part c
round(college.describe(), 2)

```

```python
# Creating a scatterplot matrix of the first ten columns of data

sns.pairplot(college.iloc[:,:10])
```

```python
# side by side boxplots of outstate versus private 

sns.boxplot(x = college.Private, y = college.Outstate)
```

```python
# Creating new variable Elite by binning the Top10perc variable 
import numpy as np

# Remember np.where?

college['Elite'] = np.where(college['Top10perc'] > 50.0, "Yes", "No")

college.Elite.head(10)
```

```python
# Getting a count of Elite and non-Elite schools
college.groupby(['Elite']).count()
```

```python
# Boxplot for side by side of Outstate vs Elite

sns.boxplot(x = college.Outstate, y = college.Elite)
```

```python
# Using histograms to investigate the quantitative variables 
# and provide insight 
# utilizing matplotlib subplotting
import matplotlib.pyplot as plt

sns.set()
fig, axes = plt.subplots(2,2, figsize=(16,8))
fig.suptitle('EDA of Quantitative Variables in the College Data Set')

# Histogram of PhD

sns.histplot(ax = axes[0,0], x = college.PhD, bins = 15)

sns.histplot(ax = axes[0,1], x = college['S.F.Ratio'], color='red', bins = 15)

sns.histplot(ax = axes[1,0], x = college.Books, color = 'green', bins = 15)

sns.histplot(ax = axes[1,1], x = college['Grad.Rate'], color = 'yellow', bins = 15)
# finish fleshing this out
```

```python
# Deriving acceptance rate from the existing data 

college['accept_rate'] = college.Accept / college.Apps

college['accept_rate'] = round(college.accept_rate, 2)

college.accept_rate.head()
```

```python
# Visualizing our new statistic 

sns.histplot(x = college.accept_rate, bins = 15)
```

```python
# getting some descriptive statistics 

college.accept_rate.describe()
```

```python
# A boxplot of elite colleges vs non and how they differ in acceptance rates 

sns.boxplot(x = college.Elite , y = college.accept_rate)
```

```python
# Making a comparison with a sarm and violin plot to get an idea of spread 

sns.set()
fig, axes = plt.subplots(1, 3, figsize=(16,8))
fig.suptitle('Comparative Visualization of Elite Colleges and Acceptance Rates')

sns.boxplot(ax = axes[0], x = college.Elite, y = college.accept_rate)

# violin plot

sns.violinplot(ax = axes[1], x = college.Elite, y = college.accept_rate)

# swarm plot 

sns.swarmplot(ax = axes[2], x = college.Elite, y = college.accept_rate)
```

```python
# Now to make up enroll rate as a type of conversion rate for these schools 

college['enroll_rate'] = college.Enroll / college.Accept

college['enroll_rate'] = round(college.enroll_rate, 2)

college.enroll_rate.head()
```

```python
# Going to look at the same plots fro enroll rate for elite and non-elite 

sns.set()
fig, axes = plt.subplots(1, 3, figsize=(16,8))
fig.suptitle('Comparative Visualization of Elite Colleges and Enrollment Rates')

sns.boxplot(ax = axes[0], x = college.Elite, y = college.enroll_rate)

# Violin plot

sns.violinplot(ax = axes[1], x = college.Elite, y = college.enroll_rate)

# Swarm plot

sns.swarmplot(ax = axes[2], x = college.Elite, y = college.enroll_rate)
```

### 9. Descriptive Statistics with the Auto Data Set 

```python
# Reading in the data set and looking at the quantitative predictors 

auto = pd.read_csv('/Users/anthonybernardi/Desktop/Werk/ISLR/Data_Sets/Auto.csv', header = 0)
```

```python
# Quantitative predictors are mpg, displacement, hp, weight, acceleration 

# Range of each qualitative predictor 

# Creating a list of lists for the quant columns

quant_vars = [auto.mpg , auto.displacement, auto.weight, auto.acceleration]

for i in quant_vars:
    print(min(i))
    print(max(i))
```

```python
# Finding the mean and the sd for each of these predictors 
# Columns are in order; mpg, displacement, weight, acceleration

for i in quant_vars:
    print("Mean: " + str(round(np.mean(i), 2)))
    print("Std. Dev: " + str(round(np.std(i), 2)))
```

```python
# Remove the 10th through 85th observations, and then compute range, standard dev, and mean

auto_dropped = auto

auto_dropped.drop(auto_dropped.index[9:86], 0, inplace=True)
```

```python
# Gut check

auto_dropped.head(15)
```

```python
# Printing out the quantitative variables again 
# Want mpg, displacement, weight, accel

quant_vars_dropped = [auto_dropped.mpg, auto_dropped.displacement, auto_dropped.weight, auto_dropped.acceleration]

for i in quant_vars_dropped:
    print(min(i))
    print(max(i))
```
