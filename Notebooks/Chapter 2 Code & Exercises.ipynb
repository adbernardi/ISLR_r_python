```python
# Changing the plot color to green
N = 100

x = np.random.rand(N)
x
y = np.random.rand(N)
print(y)

# Labeling the plot now to round it out 
plt.scatter(x, y, c='green')
plt.title('Plot of X vs Y')
plt.xlabel('This is the x-axis')
plt.ylabel('This is the y-axis')
plt.show()
```

### 1.2 Working with the Auto Data Set & EDA

```python
# Reading in our csv 
import pandas as pd 

auto = pd.read_csv('/Users/anthonybernardi/Desktop/Werk/ISLR/Data_Sets/Auto.csv', header=0)

auto.head()

# dimension equivalent 
auto.shape

# list of columns 
auto.columns
```

```python
# Basic plotting for visual EDA 

plt.scatter(auto.cylinders, auto.mpg)
plt.xlabel('Cylinders')
plt.ylabel('MPG')
plt.show()
```

```python
# trying a boxplot for what is a categorical var
import seaborn as sns 

sns.boxplot(x = auto.cylinders, y = auto.mpg)
```

```python
# Trying the same with a histogram -- seaborn is the GOAT

sns.histplot(x = auto.mpg)
```

```python
# Customizing the histogram, adjusting bins, etc 

sns.histplot(x = auto.mpg, color = "red")
```

```python
# Adding binz 

sns.histplot(x = auto.mpg, color = "red", bins = 15)
```

```python
# Pair plots!

sns.pairplot(auto)
```
