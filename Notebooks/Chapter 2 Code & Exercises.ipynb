```python
# Taking R's rnorm() and cor() functions and translating them
import numpy as np 
import matplotlib.pyplot as plt 

N = 100

x = np.random.rand(N)
x
y = np.random.rand(N)
print(y)

# Labeling the plot now to round it out 
plt.scatter(x, y)
plt.title('Plot of X vs Y')
plt.xlabel('This is the x-axis')
plt.ylabel('This is the y-axis')
plt.show()
```

```python
# Changing the plot color to green
N = 100

x = np.random.rand(N)
x
y = np.random.rand(N)
print(y)

# Labeling the plot now to round it out 
plt.scatter(x, y, c='green')
plt.title('Plot of X vs Y')
plt.xlabel('This is the x-axis')
plt.ylabel('This is the y-axis')
plt.show()
```
